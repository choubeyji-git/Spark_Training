```
# a)Create HR_Employee Pysaprk dataFrame. 
HR_df=spark.read.csv("dbfs:/FileStore/tables/HR_Employee.csv",header= True ,inferSchema=True)
HR_df.show()
```

```
#b)	Show printSchema & show 10 rows of Dataset
HR_df.printSchema()
HR_df.show(10)
```

```
HR_df.createOrReplaceGlobalTempView("employeetable")
```

```
#c)	Most employee is working in which department
spark.sql("select Department ,count(*) as most_nu_employee from employeetable group by Department order by most_nu_employee desc ").show()
```

```
# d)Highest number of job roles
spark.sql("select Max(JobRole) as Job_role,count(*) as most_nu_employee from employeetable ").show()
```

```
# e)    Which gender have higher strength as workforce?
spark.sql("select Max(Gender) as Gender,count(*) as most_Strength from employeetable").show()
```

```
# f) Compare the marital status of employee and find the most frequent status.
spark.sql("select MaritalStatus as Gender,count(*) as most_frequent_status from employeetable where BusinessTravel ='Travel_Frequently' group by MaritalStatus ").show()
```

```
# g)    Mostly hired employee have qualification
spark.sql("select Education, count(*) as Number_Employee from employeetable  group by Education order by Number_Employee desc").show()
```

```
# h)    Find the count of employee from which education fields
spark.sql("select EducationField, count(*) as Number_Employee from employeetable group by EducationField order by Number_Employee desc").show()
```

```
# i) What is the job satisfaction level of employee?
spark.sql("select    JobSatisfaction , count(*) as Number_Employee from employeetable  group by JobSatisfaction order by Number_Employee desc").show()
```

```
# j)    Does most of employee do overtime: Yes or No?
spark.sql("select Overtime, count(*) as Number_Employee from employeetable where OverTime ='Yes' or Overtime ='No' group by Overtime ").show()
```

```
# k)    Find Min & Max Salaried employees.
spark.sql("select Max(Income) as Maximun_salary,Min(Income) as Minimun_salary  from employeetable ").show()
```

```
# l) Does most of the employee do business travel? Find of the employees counts for each category   BusinessTravel   
spark.sql("select JobRole , count(*) as Number_Employee from employeetable where BusinessTravel ='Travel_Frequently' group by JobRole ").show()
```

```
# m)    Find the AVG salary of graduate employee.
spark.sql("select mean(Income) as Averge_salary_employee  from employeetable where Education = 'Bachelor' ").show()
```

```
#n)	Find the employee qualification receiving salary lower than equal to avg. salary of all employee. 
spark.sql("select Income, Education from employeetable where Income <= (select avg(Income) from employeetable)").show()
```

```
#o)	When does the employee have highest chance of promotion in terms of working year
spark.sql("select Performance_Rating, count(*) as PromotionChance from employeetable group by Performance_Rating order by PromotionChance desc ").show()
```

```
#p)	Highest attrition is in which department?
spark.sql("select Department, count(*) as totalcount from employeetable where Attrition='Yes' group by Department order by totalcount desc ").show()
```

```
#q)	Show marital status of Person having highest attrition rate
spark.sql("select MaritalStatus, count(*) as HighAttrition from employeetable where Attrition='Yes' group by MaritalStatus order by HighAttrition desc ").show(1)
```
