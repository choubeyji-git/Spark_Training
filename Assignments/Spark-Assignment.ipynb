```
#g)	Show Analysis for each month with total number of cancellations. (Descending).
spark.sql("select MONTH, sum(CANCELLED) as TOTALCANCELLED from datanewtable group by MONTH order by TOTALCANCELLED desc").show()
```

```
#i)	Show the airlines that make the maximum number of cancellations.
flights_delay_df=spark.sql("select AIRLINE,count(*) as CANCELLED from datanewtable group by AIRLINE order by CANCELLED desc")
flights_delay_df.show()
```

```
#j)	Find and order airlines in descending that make the most number of diversions.
flights_delay_df=spark.sql("select AIRLINE,count(*) as DIVERTED from datanewtable group by AIRLINE order by DIVERTED desc")
flights_delay_df.show()
```

```
#k)	Show days of month that seen the most number of diversion.
flights_delay_df=spark.sql("select MONTH,count(*) as DIVERTED from datanewtable group by MONTH order by DIVERTED desc")
flights_delay_df.show()
```

```
#l)	Calculate mean and standard deviation of departure delay for all flights in minutes.
spark.sql("select AIRLINE, AVG(DEPARTURE_DELAY) as AvgDepDelay, STDDEV(DEPARTURE_DELAY) as devDepDelay from datanewtable group by AIRLINE ").show()
```

```
#m)	Calculate mean and standard deviation of arrival delay for all flights in minutes.
spark.sql("select AIRLINE, AVG(ARRIVAL_DELAY) as AvgArrDelay, STDDEV(ARRIVAL_DELAY) as ArrDepDelay from datanewtable group by AIRLINE ").show()
```

```
#n)	Find all diverted Route from a source to destination Airport & which route is the most diverted.
flights_delay_df=spark.sql("select AIRLINE,count(*) as DIVERTED from datanewtable group by AIRLINE order by DIVERTED desc")
flights_delay_df.show()
```
